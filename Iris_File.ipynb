```python
import matplotlib.pyplot as plt

months=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption=[1200,1300,1100,1500,1400,1600]

plt.plot(months, energy_consumption, marker='o', color='b', linestyle='--')

plt.title('Energy Consumption Over Months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (kWh)')

plt.grid(True)

plt.show
```

```python
energy_sources=['Solar','Wind','Hydropower','Biomass']
energy_consumption=[1000,800,1200,900]

plt.bar(energy_sources, energy_consumption, color=['b','g','r','y'])

plt.title('Energy Consumption by Source')
plt.xlabel('Energy Source')
plt.ylabel('Energy Consumption (kWh)')

plt.show()
```

```python
plt.pie(energy_consumption, labels=energy_sources, autopct='%1.1f%%', startangle=90)

plt.title('Energy Consumption by Source')

plt.show()
```

```python
# Sample data for energy consumption and carbon emissions

carbon_emissions = [200, 500, 450, 300] # in kg CO2

#Create a scatter plot

plt.scatter(energy_consumption, carbon_emissions, color='red')

#Add titles and Labels

plt.title('Energy Consumption vs Carbon Emissions')

plt.xlabel('Energy Consumption (MWh)')

plt.ylabel('Carbon Emissions (kg CO2)')

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data={
    "Solar" : [1000,800,1200,900],
    "Wind" : [1200,1300,1100,1500],
    "Hydropower" : [2900, 3100,3000, 2800],
    "Biomass" : [2500, 2700,2500, 2400]
}

df=pd.DataFrame(data)

sns.pairplot(df)

plt.show()
```

```python
correlation_matrix=df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidth=0.5)

plt.title('Correlation Matrix')
plt.show()
```

```python
sns.boxplot(data=df)

plt.title('Box Plot')
plt.show()
```

```python
sns.violinplot(data=df)

plt.title('Violin Plot')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')


# Display the first few rows of the dataset
print(iris.head())

# 1. Pair Plot
sns.pairplot(iris, hue='species', diag_kind='kde')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)  # Add a title
plt.show()

# 2. Heatmap
correlation_matrix = iris.drop('species', axis=1).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Features')
plt.show()

# 3. Box Plot
plt.figure(figsize=(10, 6))  # Adjust figure size
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Box Plot of Sepal Length by Species')
plt.show()

# 4. Violin Plot
plt.figure(figsize=(10, 6))  # Adjust figure size
sns.violinplot(data=iris, x='species', y='petal_width')
plt.title('Violin Plot of Petal Width by Species')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

iris = sns.load_dataset('iris')

iris_encoded = iris.copy()

le = LabelEncoder()

iris_encoded['species'] = le.fit_transform(iris_encoded['species'])

print(iris_encoded.head())

correlation_matrix = iris_encoded.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Features and Species')
plt.show()
```
